````python
Disenority = dict.fromkeys(('Junior','Semisenior','Senior'),0)
for num in Senor:
    if num <= 2 in Senor:
        Disenority['Junior'] += 1
    elif 2 < num <= 5 in Senor:
        Disenority['Semisenior'] += 1
    else:
        Disenority['Senior'] += 1
````
